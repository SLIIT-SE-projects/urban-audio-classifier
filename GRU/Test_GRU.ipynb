```python
import IPython.display as ipd
import librosa
import librosa.display
import pandas as pd
import numpy as np
import tensorflow as tf
import joblib
```

```python
# Load the GRU model and its specific scaler
model = tf.keras.models.load_model('./best_gru_model.keras')
scaler = joblib.load('./standard_scaler_gru.pkl')
```

```python
classes=['air_conditioner', 'car_horn', 'children_playing', 'dog_bark', 'drilling', 'engine_idling', 'gun_shot', 'jackhammer', 'siren', 'street_music']
```

```python
def extract_features(file_path):
    # Fixed variable name from audio_path to file_path
    X, sample_rate = librosa.load(file_path)
    mels = np.mean(librosa.feature.melspectrogram(y=X, sr=sample_rate).T, axis=0)
    return mels
```

```python
def predict(file_path, classes):
    # Extract features from the file
    mels = extract_features(file_path)
    
    # Scale the features
    mels = scaler.transform(mels.reshape(1,-1))
    
    # Reshape features to the required input shape for the GRU model (1, 1, 128)
    mels = mels.reshape(1, 1, 128)
    
    # Get the prediction from the model
    prediction = model.predict(mels)
    
    # Get the index of the maximum value from the prediction
    predicted_index = np.argmax(prediction, axis=1)[0]
    
    # Return the class name with the predicted index
    return classes[predicted_index]
```

## Test 1: Siren

```python
audio_path='../sounds/22601-8-0-9.wav'
ipd.Audio(filename=audio_path)
```

```python
class_name = predict(audio_path, classes)
print("Predicted class:", class_name)
```
